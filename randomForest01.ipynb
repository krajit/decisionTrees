# Decision trees

```python
import pandas as pd
import numpy as np
from sklearn import tree

df = pd.DataFrame()
n = 2000 # num data
df['x1'] = [np.random.randint(0,100) for i in range(n)]
df['x2'] = [np.random.randint(0,100) for i in range(n)]

df['y'] = [1 if (df['x1'][i]  >= df['x2'][i] ) else 0 for i in range(n)]

x = df.drop(columns=['y'])
x = np.array(x)
y = df['y']
y = np.array(y)

import matplotlib.pyplot as plt

plt.plot(x[:, 0][y==0], x[:, 1][y==0], "yo", marker='.')
plt.plot(x[:, 0][y==1], x[:, 1][y==1], "bo", marker='.')
```

```python
clf = tree.DecisionTreeClassifier()#max_depth=3,)
clf.fit(x,y)
```

```python
tree.plot_tree(clf)
```

```python
y1 = clf.predict(x)
```

```python
y1
```

```python
len(y1)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y, y1)
conf_matrix
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y, y1).round(2)
```

```python
plt.plot(x[:, 0][y1==0], x[:, 1][y1==0], "yo", marker='.')
plt.plot(x[:, 0][y1==1], x[:, 1][y1==1], "bo", marker='.')
```

# Randome Forests

```python
import pandas as pd
import numpy as np
from sklearn import tree

df = pd.DataFrame()
n = 200 # num data
df['x1'] = [np.random.randint(0,100) for i in range(n)]
df['x2'] = [np.random.randint(0,100) for i in range(n)]

df['y'] = [1 if (df['x1'][i]  >= df['x2'][i] ) else 0 for i in range(n)]

x = df.drop(columns=['y'])
x = np.array(x)
y = df['y']
y = np.array(y)

import matplotlib.pyplot as plt

plt.plot(x[:, 0][y==0], x[:, 1][y==0], "yo", marker='.')
plt.plot(x[:, 0][y==1], x[:, 1][y==1], "bo", marker='.')
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=3)
clf.fit(x, y)
```

```python
from sklearn.metrics import confusion_matrix
y1 = clf.predict(x)
conf_matrix = confusion_matrix(y, y1)
conf_matrix
```
