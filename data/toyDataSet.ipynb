# Data : y = 1 if x1 <= x0

```python
import pandas as pd
import numpy as np
from sklearn import tree

df = pd.DataFrame()
n = 2000 # num data
df['x1'] = [np.random.randint(0,100) for i in range(n)]
df['x2'] = [np.random.randint(0,100) for i in range(n)]

df['y'] = [1 if (df['x1'][i]  >= df['x2'][i] ) else 0 for i in range(n)]

x = df.drop(columns=['y'])
x = np.array(x)
y = df['y']
y = np.array(y)

import matplotlib.pyplot as plt

plt.plot(x[:, 0][y==0], x[:, 1][y==0], "yo", marker='.')
plt.plot(x[:, 0][y==1], x[:, 1][y==1], "bo", marker='.')
```
